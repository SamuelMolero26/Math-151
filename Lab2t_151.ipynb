```python
t = Symbol('t')
x1 = (3*a*t)/(1+(t**3))
y1 = (3*a*(t**2))/(1+(t**3))
x1.subs(a,2)
y1.subs(a,2)
plot1 = plot_parametric(x1,y1,(t,0,6),show=False)
plot1.show()

```
