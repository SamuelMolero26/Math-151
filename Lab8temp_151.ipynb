#### 1b

```python
#start code here
df=diff(f,x)
cvals=solve(df,x)
cvals_float=[i.evalf() for i in cvals]
print('The critical values are',cvals_float)
plot(df,(x,-1,2),ylim=(-1,1))
endpts=[-oo,-.23,0,1.62,oo] 
testvals=[-.24,-.1,1.60,1.65] 
for n in range(4): 
    if df.subs(x,testvals[n])>0:
        print('f is increasing on (',endpts[n],',',endpts[n+1],')')
    else:
        print('f is decreasing on (',endpts[n],',',endpts[n+1],')')

```

```python
ddf=diff(f,x,2)
print('The second derivative is',ddf.simplify())
c2vals=solve(ddf,x)
c2vals_float=[i.evalf() for i in c2vals]
print('The 2nd derivative critical values are',c2vals_float)
endpts=[-oo,-.12,0,1.24,oo]
testvals=[-.15,-.1,1.2,1.3] # NOTICE now we have elements [0] through [5] in this list
for n in range(4):
    if ddf.subs(x,testvals[n])>0:
        print('f is concave up on (',endpts[n],',',endpts[n+1],')')
    else:
        print('f is concave down on (',endpts[n],',',endpts[n+1],')')

```

```python
#start code here
x=symbols('x')
f=(1+x)/(1+(x**2))
difOne=diff(f,x)
difTwo=diff(difOne,x)
inflection=solve(difTwo,x)
slope=(((1+inflection[1])/(1+inflection[1]**2))-((1+inflection[0])/(1+inflection[0]**2)))/((inflection[1]-inflection[0]))
yint=(1+inflection[0])/(1+inflection[0]**2)-slope*inflection[0];
line=(slope*x).evalf()+yint.evalf()
print("The equation of the line is",line)
fplot=plot(f,line,(x,-10,5))
```
